# **Chapter 3 Boundary Value Problems: Diskritisasi**

## **3.1 Model Problems**

Masalah nilai batas (BVP) paling sederhana yang akan kita temui adalah versi satu dimensi dari persamaan Poisson

$ u''(x) = f(x)$

Biasanya kita menyelesaikan persamaan ini pada interval hingga dengan kondisi batas Dirichlet atau Neumann. Karena terdapat dua turunan dalam persamaan tersebut, kita memerlukan dua kondisi batas untuk menyelesaikan PDE (sebenarnya ODE dalam kasus ini) secara unik. Untuk memulai, mari kita pertimbangkan masalah dasar berikut

$ u''(x) = f(x)    Ω = [a,b] $

$ u(a) = α         u(b) = β $.

Masalah-masalah BVP semacam ini sering kali muncul sebagai hasil dari melihat bentuk keadaan tunak (steady-state) dari suatu PDE yang bergantung pada waktu. Sebagai contoh, jika kita mempertimbangkan solusi keadaan tunak dari persamaan panas

Persamaan panas satu dimensi dengan sumber termal:

$$
u_t(x,t) = \kappa u_{xx}(x,t) + \Psi(x,t) \quad \Omega = [0,T] \times [a,b]
$$

dengan kondisi awal dan batas:

$$
u(x,0) = u^0(x) \quad u(a,t) = \alpha(t) \quad u(b,t) = \beta(t)
$$

kita akan menyelesaikan persamaan tersebut di mana $u_t = 0$ dan sampai pada

$$
u''(x) = -\frac{\Psi}{\kappa},
$$

sebuah versi dari persamaan Poisson di atas.

Dalam dimensi spasial yang lebih tinggi, turunan kedua berubah menjadi Laplacian. Notasi bervariasi tergantung penulis tetapi semua pernyataan berikut ini adalah ekuivalen:

$$
\nabla^2 u(\vec{x}) = f(\vec{x})
$$

$$
\Delta u(\vec{x}) = f(\vec{x})
$$

$$
\Delta u(\vec{x}) = f(\vec{x})
$$

$$
\sum_{i=1}^N u_{x_i x_i} = f(\vec{x}).
$$

## **3.2 One-Dimensional Discretization**

Sebagai pendekatan pertama untuk menyelesaikan persamaan Poisson satu dimensi, mari kita pecah domain menjadi $m$ titik, yang sering disebut *mesh* atau *grid*. Tujuan kita adalah mengaproksimasi fungsi tak diketahui $u(x)$ sebagai titik-titik mesh $x_i$. Pertama kita dapat menghubungkan jumlah titik mesh $m$ dengan jarak antar titik dengan

$$
\Delta x = \frac{1}{m + 1}.
$$

Titik-titik mesh $x_i$ dapat ditulis sebagai

$$
x_i = a + i \Delta x.
$$

Kita dapat membiarkan $\Delta x$ bervariasi dan banyak dari rumus-rumus di atas hanya memerlukan modifikasi kecil, tetapi kita akan meninggalkan hal itu untuk pekerjaan rumah.

Secara notasi kita juga akan mengadopsi notasi

$$
U_i \approx u(x_i)
$$

sehingga $U_i$ adalah solusi aproksimasi pada titik-titik grid dan kita mempertahankan huruf kecil $u$ untuk menunjukkan solusi sebenarnya.

Untuk menyederhanakan diskusi kita, mari kita pertimbangkan ODE

$$
u''(x) = f(x) \quad \Omega = [0,1]
$$
$$
u(0) = \alpha \quad u(1) = \beta.
$$

Menerapkan aproksimasi beda terpusat orde kedua untuk turunan kedua, kita memperoleh persamaan

$$
D^2 U_i = \frac{1}{\Delta x^2} (U_{i+1} - 2U_i + U_{i-1})
$$

sehingga kita berakhir dengan ekspresi aljabar aproksimasi pada setiap titik grid berupa

$$
\frac{1}{\Delta x^2} (U_{i+1} - 2U_i + U_{i-1}) = f(x_i) \quad i = 1,2,3,\dots,m.
$$

Perhatikan pada titik ini bahwa persamaan-persamaan aljabar ini saling terkait (coupled) karena setiap $U_i$ bergantung pada tetangganya. Hal ini berarti kita dapat menuliskannya sebagai sistem persamaan yang saling terkait

$$
AU = F.
$$

Tuliskan sistem persamaannya

**Perhatikan kondisi batasnya!**

$$
\frac{1}{\Delta x^2}
\begin{bmatrix}
-2 & 1 & & & \\
1 & -2 & 1 & & \\
& 1 & -2 & 1 & \\
& & 1 & -2 & 1 \\
& & & 1 & -2
\end{bmatrix}
\begin{bmatrix}
U_1 \\
U_2 \\
U_3 \\
U_4 \\
U_5
\end{bmatrix}
=
\begin{bmatrix}
f(x_1) - \frac{\alpha}{\Delta x^2} \\
f(x_2) \\
f(x_3) \\
f(x_4) \\
f(x_5) - \frac{\beta}{\Delta x^2}
\end{bmatrix}.
$$

## Contoh

Ingin menyelesaikan BVP

$$
u_{xx} = e^x, \quad x \in [0,1] \quad \text{dengan} \quad u(0) = 0, \quad \text{dan} \quad u(1) = 3
$$

melalui pembentukan sistem persamaan linear.

$$
u_{xx} = e^x
$$

$$
u_x = A + e^x
$$

$$
u = Ax + B + e^x
$$

$$
u(0) = B + 1 = 0 \quad \Rightarrow \quad B = -1
$$

$$
u(1) = A - 1 + e^1 = 3 \quad \Rightarrow \quad A = 4 - e
$$

$$
u(x) = (4 - e)x - 1 + e^x
$$

```python
# Pengaturan masalah (Problem setup)
a = 0.0                  # Batas kiri domain
b = 1.0                  # Batas kanan domain
u_a = 0.0                # Nilai kondisi batas di x = a  (Dirichlet: u(a) = 0)
u_b = 3.0                # Nilai kondisi batas di x = b  (Dirichlet: u(b) = 3)
f = lambda x: numpy.exp(x)                      # Fungsi sumber / right-hand side: f(x) = e^x
u_true = lambda x: (4.0 - numpy.exp(1.0)) * x - 1.0 + numpy.exp(x)   # Solusi analitik eksak untuk perbandingan

# Diskritisasi (Discretization)
m = 10                   # Jumlah interval di dalam domain (jumlah titik interior = m)
x_bc = numpy.linspace(a, b, m + 2)   # Titik-titik grid termasuk batas (m+2 titik)
x = x_bc[1:-1]           # Titik-titik interior saja (tanpa batas), ukuran m
delta_x = (b - a) / (m + 1)   # Jarak antar titik grid (lebar sel)

# Membuat matriks A (matriks koefisien sistem linear)
A = numpy.zeros((m, m))       # Matriks m × m diisi nol terlebih dahulu
diagonal = numpy.ones(m) / delta_x**2   # Nilai dasar 1/h²

A += numpy.diag(diagonal * -2.0, 0)     # Elemen diagonal utama: -2/h²
A += numpy.diag(diagonal[:-1], 1)       # Elemen superdiagonal (atas diagonal): 1/h²
A += numpy.diag(diagonal[:-1], -1)      # Elemen subdiagonal (bawah diagonal): 1/h²

# Membuat vektor sisi kanan (right-hand side / RHS)
b = f(x)                        # f(x_i) untuk setiap titik interior
b[0] -= u_a / delta_x**2        # Koreksi kondisi batas kiri: -u(a)/h² pada baris pertama
b[-1] -= u_b / delta_x**2       # Koreksi kondisi batas kanan: -u(b)/h² pada baris terakhir

# Menyelesaikan sistem linear A U = b
U = numpy.empty(m + 2)          # Array lengkap termasuk titik batas (ukuran m+2)
U[0] = u_a                      # Terapkan kondisi batas kiri
U[-1] = u_b                     # Terapkan kondisi batas kanan
U[1:-1] = numpy.linalg.solve(A, b)   # Solusi numerik di titik-titik interior

# Membuat plot hasil
fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)
axes.plot(x_bc, U, 'o', label="Computed")           # Titik-titik hasil komputasi (numerik)
axes.plot(x_bc, u_true(x_bc), 'k', label="True")    # Kurva solusi analitik eksak
axes.set_title("Solution to $u_{xx} = e^x$")        # Judul plot
axes.set_xlabel("x")                                # Label sumbu x
axes.set_ylabel("u(x)")                             # Label sumbu y
plt.show()                                          # Tampilkan plot
```
